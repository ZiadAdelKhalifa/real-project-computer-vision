```python
import cv2
import mediapipe as mp

img = cv2.imread('salah.jpg')

mp_face_mesh = mp.solutions.face_mesh.FaceMesh()

img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

result = mp_face_mesh.process(img_rgb)

h, w, _ = img.shape

if result.multi_face_landmarks:
    for facial_landmarks in result.multi_face_landmarks:
        for i in range(0, 468):
            pt = facial_landmarks.landmark[i]
            x = int(pt.x * w)
            y = int(pt.y * h)
            cv2.circle(img, (x, y), 2, (100, 100, 0), -1)

cv2.imshow('img', img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2
import mediapipe as mp

# Initialize MediaPipe Face Mesh
mp_face_mesh = mp.solutions.face_mesh
face_mesh = mp_face_mesh.FaceMesh()

# Open webcam
cap = cv2.VideoCapture(0)  # 0 corresponds to the default camera (you can change it if you have multiple cameras)

# Get the video dimensions and create a VideoWriter object
width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
out = cv2.VideoWriter('output_video.avi', cv2.VideoWriter_fourcc('M','J','P','G'), 10, (width, height))

while cap.isOpened():
    # Read a frame from the webcam
    ret, frame = cap.read()
    if not ret:
        break

    # Convert the frame to RGB
    frame_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)

    # Process the frame with Face Mesh
    result = face_mesh.process(frame_rgb)

    # Draw landmarks on the face
    if result.multi_face_landmarks:
        for facial_landmarks in result.multi_face_landmarks:
            for i in range(0, 468):
                pt = facial_landmarks.landmark[i]
                x = int(pt.x * frame.shape[1])
                y = int(pt.y * frame.shape[0])
                cv2.circle(frame, (x, y), 2, (100, 100, 0), -1)

    # Write the frame to the output video
    out.write(frame)

    # Display the frame
    cv2.imshow('Face Mesh', frame)

    # Break the loop if 'q' key is pressed
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release the webcam, video writer, and close all windows
cap.release()
out.release()
cv2.destroyAllWindows()
```
