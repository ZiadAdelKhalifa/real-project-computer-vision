```python
import cv2
import numpy as np
from google.colab import files

# Upload the video file manually
uploaded = files.upload()

# Assuming you've uploaded the video as 'walkers.mp4'
video_path = 'walkers.mp4'

cap = cv2.VideoCapture(video_path)

frame_wid = int(cap.get(3))
frame_hig = int(cap.get(4))

fourcc = cv2.VideoWriter_fourcc('X', 'V', 'I', 'D')
out = cv2.VideoWriter('output.mp4', fourcc, 30, (frame_wid, frame_hig))  # Use the original frame size

ret, frame1 = cap.read()
ret, frame2 = cap.read()

while ret:  # Check if frames are successfully read
    diff = cv2.absdiff(frame1, frame2)
    gray = cv2.cvtColor(diff, cv2.COLOR_BGR2GRAY)
    blur = cv2.GaussianBlur(gray, (5, 5), 0)
    _, thresh = cv2.threshold(blur, 60, 255, cv2.THRESH_BINARY)
    dilated = cv2.dilate(thresh, None, iterations=10)
    contours, _ = cv2.findContours(dilated, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)

    for contour in contours:
        (x, y, w, h) = cv2.boundingRect(contour)
        if cv2.contourArea(contour) < 600:
            continue
        cv2.rectangle(frame1, (x, y), (x + w, y + h), (0, 255, 0), 2)
        cv2.putText(frame1, 'Statues:{}'.format('Movement'), (10, 20), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 0, 255), 3)

    # Resize the frame after processing and before writing to the output video
    image = cv2.resize(frame1, (frame_wid, frame_hig))
    out.write(image)

    frame1 = frame2
    ret, frame2 = cap.read()

cv2.destroyAllWindows()
cap.release()
out.release()

# Download the processed video
files.download('output.mp4')
```
